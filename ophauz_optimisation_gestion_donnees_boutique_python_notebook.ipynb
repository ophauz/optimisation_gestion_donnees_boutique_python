### 4.5. On approfondit avec l'approche interquartile

Avec l'approche interquartile, est considérée comme outlier toute valeur extrême, supérieure à Q3 ou inférieure à Q1 de 1,5 fois l’espace inter-quartile.

```python
# on définit la variable Q1, le premier quartile (25% des valeurs en dessous)
Q1 = np.percentile(final_outliers['price'], 25)

# on définit la variable Q3, le troixième quartile (75% des valeurs en dessous)
Q3 = np.percentile(final_outliers['price'], 75)

# on définit IQR, l'espace inter-quartile
IQR = Q3 - Q1

# on définit les outliers 'upper' comme étant supérieurs à Q3 de 1.5 fois l'IQR
upper = final_outliers['price'] >= (Q3 + 1.5 * IQR)

# on affiche l'index des produits dont le prix est un outlier
print(np.where(upper))
```

```python
# on affiche le nombre d'outliers
print(np.count_nonzero(upper))
```

```python
# on affiche les outliers, le plus bas d'entre eux est surligné en vert
final_outliers.loc[upper].style.highlight_min(color = 'lightgreen', axis = 0, subset = 'price') 
```

Avec la méthode interquartile, il y a :

- 32 valeurs qui pourraient être qualifiées d'atypiques

- avec une valeur minimale pour 'price' de 85.6

### 4.6. Scatter plot et visualisation des outliers

```python
# on définit les outliers comme toute valeur de 'price' supérieure à 83.7
outliers = final_outliers[final_outliers.price.gt(85.6)]

# on trace le scatterplot des prix pour chaque product_id
ax = final_outliers.plot.scatter(x = 'price', y = 'product_id', figsize = (15, 5), s = 30, alpha = 0.5)

# on affiche les outliers en rouge
outliers.plot.scatter(x = 'price', y = 'product_id', color = 'red', ax = ax, label = 'outliers', s = 30, alpha = 0.5)
```

### 4.7. Calcul du Z-score

Le score Z correspond à l'expression d'un écart à la moyenne exprimé en fraction d'écart-type.

La distribution des scores est « centrée réduite » et chaque score Z exprime la distance de l'observation à la moyenne en nombre d’écart-type.

On pose alors un score seuil au delà duquel l'observation est jugée comme outlier.

```python
# on calcule le z-score de chaque article
z_sc = np.abs(stats.zscore(final_outliers['price']))

# on affiche le z-score de chaque article
print(z_sc)
```

```python
# on détermine le seuil pour décider quelle observation doit être considérée comme extrême
# moyenne + (1.96 x écart-type)
# on se situe dans les 2,5% les plus faibles

seuil = 1.96

# on affiche l'index des produits dont le prix est un outlier

print(np.where(z_sc > 1.96))
```

```python
# on affiche le nombre d'outliers
print(np.count_nonzero(z_sc > 1.96))
```

```python
# on affiche les outliers, le plus bas d'entre eux est surligné en vert
final_outliers.loc[z_sc > 1.96].style.highlight_min(color = 'lightgreen', axis = 0, subset = 'price')
```

Avec la méthode du Score Z, il y a :

- 30 valeurs qui pourraient être qualifiées d'atypiques 

- avec une valeur minimale pour 'price' de 88.4
