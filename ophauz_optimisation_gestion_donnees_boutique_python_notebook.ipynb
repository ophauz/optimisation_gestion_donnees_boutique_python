### 4.4. On utilise les bo√Ætes √† moustache des librairies matplotlib et seaborn pour afficher les valeurs atypiques

Une valeur atypique ou extr√™me (outlier) est une observation ‚Äúqui semble d√©vier notoirement des autres √©l√©ments de l‚Äô√©chantillon auquel elle appartient‚Äù (Grubbs, 1969)

### 4.5. On approfondit avec l'approche interquartile

Avec l'approche interquartile, est consid√©r√©e comme outlier toute valeur extr√™me, sup√©rieure √† Q3 ou inf√©rieure √† Q1 de 1,5 fois l‚Äôespace inter-quartile.

```python
# on d√©finit la variable Q1, le premier quartile (25% des valeurs en dessous)
Q1 = np.percentile(final_outliers['price'], 25)

# on d√©finit la variable Q3, le troixi√®me quartile (75% des valeurs en dessous)
Q3 = np.percentile(final_outliers['price'], 75)

# on d√©finit IQR, l'espace inter-quartile
IQR = Q3 - Q1

# on d√©finit les outliers 'upper' comme √©tant sup√©rieurs √† Q3 de 1.5 fois l'IQR
upper = final_outliers['price'] >= (Q3 + 1.5 * IQR)

# on affiche l'index des produits dont le prix est un outlier
print(np.where(upper))
```

```python
# on affiche le nombre d'outliers
print(np.count_nonzero(upper))
```

```python
# on affiche les outliers, le plus bas d'entre eux est surlign√© en vert
final_outliers.loc[upper].style.highlight_min(color = 'lightgreen', axis = 0, subset = 'price') 
```

Avec la m√©thode interquartile, il y a :

- 32 valeurs qui pourraient √™tre qualifi√©es d'atypiques

- avec une valeur minimale pour 'price' de 85.6

### 4.6. Scatter plot et visualisation des outliers

```python
# on d√©finit les outliers comme toute valeur de 'price' sup√©rieure √† 83.7
outliers = final_outliers[final_outliers.price.gt(85.6)]

# on trace le scatterplot des prix pour chaque product_id
ax = final_outliers.plot.scatter(x = 'price', y = 'product_id', figsize = (15, 5), s = 30, alpha = 0.5)

# on affiche les outliers en rouge
outliers.plot.scatter(x = 'price', y = 'product_id', color = 'red', ax = ax, label = 'outliers', s = 30, alpha = 0.5)
```

### 4.7. Calcul du Z-score

Le score Z correspond √† l'expression d'un √©cart √† la moyenne exprim√© en fraction d'√©cart-type.

La distribution des scores est ¬´ centr√©e r√©duite ¬ª et chaque score Z exprime la distance de l'observation √† la moyenne en nombre d‚Äô√©cart-type.

On pose alors un score seuil au del√† duquel l'observation est jug√©e comme outlier.

```python
# on calcule le z-score de chaque article
z_sc = np.abs(stats.zscore(final_outliers['price']))

# on affiche le z-score de chaque article
print(z_sc)
```

```python
# on d√©termine le seuil pour d√©cider quelle observation doit √™tre consid√©r√©e comme extr√™me
# moyenne + (1.96 x √©cart-type)
# on se situe dans les 2,5% les plus faibles

seuil = 1.96

# on affiche l'index des produits dont le prix est un outlier

print(np.where(z_sc > 1.96))
```

```python
# on affiche le nombre d'outliers
print(np.count_nonzero(z_sc > 1.96))
```

```python
# on affiche les outliers, le plus bas d'entre eux est surlign√© en vert
final_outliers.loc[z_sc > 1.96].style.highlight_min(color = 'lightgreen', axis = 0, subset = 'price')
```

Avec la m√©thode du Score Z, il y a :

- 30 valeurs qui pourraient √™tre qualifi√©es d'atypiques 

- avec une valeur minimale pour 'price' de 88.4
